```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
weather = pd.read_csv('https://coded2.herokuapp.com/datavizpandas/london2018.csv')
weather
```

```python
#Line graph is default
#plotting temperature using line graph(default)
weather.plot(y="Tmax", x="Month")
```

```python
weather.plot(x="Month",y=["Tmax","Tmin"])
```

```python
#barchart
weather.plot.bar(x="Month",y="Tmax")
plt.savefig("bar.png")
```

```python
#horizontal bar chart
weather.plot.barh(x="Month",y="Tmax")
plt.savefig("barh.png")
```

```python
weather.plot.bar(x="Month",y=["Tmax","Tmin"])
```

```python
#Multiple charts
#If we want to create a set of separate charts for each type of data
weather.plot.bar(x="Month",y=["Tmax","Tmin","Rain","Sun"],subplots=True,layout=(2,2))
plt.savefig("multibar.png")
```

```python
#Scatter Plot
#A scatter plot plots a series of points that correspond to two variables 
#allows us to determine if there is a relationship between them.
weather.plot.scatter(x="Sun",y="Rain")
```

```python
#pie_chart
meals = pd.read_csv("https://coded2.herokuapp.com/datavizpandas/meals2.csv", index_col = 0)
meals
```

```python
meals.plot.pie(subplots=True)
plt.savefig("pie.png")
```

```python
#Histogram
weather.plot.hist(y="Rain",bins=20)
```

```python
#Boxplot
weather.plot.box(y="Tmax")
```

```python
#Confirming the boxplot 
weather["Tmax"].describe()
```

```python
weather.plot.box(y=["Tmax","Tmin"],subplots=True,layout=(1,2),figsize=(10,5))
```
